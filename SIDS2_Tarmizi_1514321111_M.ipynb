```python
from mpl_toolkits.basemap import Basemap
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(1, figsize=(19,13))

# Sample it down to only the Java Island region
lon_min, lon_max = 105.21, 114.60527777778
lat_min, lat_max = -8.78, -5.8758333333333

# Mercator of Java Island
m1 = Basemap(projection='merc',
             llcrnrlat=lat_min,
             urcrnrlat=lat_max,
             llcrnrlon=lon_min,
             urcrnrlon=lon_max)

m1.drawmapboundary(fill_color='#A6CAE0', linewidth=0)
m1.fillcontinents(color='green', alpha=0.7, lake_color='grey')
m1.drawcoastlines(linewidth=0.1, color="white")

lons = [ 111.3893264,111.3387387,112.561742,111.0225624,112.4217213,110.3398252,108.6952968,107.5731164,106.863809,105.3577406]
lats = [ -7.9708679,-8.1353861,-7.4548864,-8.1015432,-7.0247246, -7.803249,-7.4616468,-6.9034443,-6.7340184,-6.6188681,]
labels = ['ponorogo', 'Trenggalek', 'Malang', 'Pacitan','Mojokerto','Yogyakarta','Cilacap','Bandung','Cianjur','Pandeglang']

x_offsets = [10000, -30000, -25000]
y_offsets = [5000, -25000, -25000]

x,y = m1(lons, lats)
m1.plot(x, y, 'bo', markersize=10)

plt.title("Pemetaan Kota di Pulau Jawa")

for label, x, y in zip(labels, x, y):
    plt.text(x+10000, y+5000, label)
    
plt.show()
```
